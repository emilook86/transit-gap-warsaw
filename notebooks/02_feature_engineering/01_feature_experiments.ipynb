# Feature Engineering Experiments

**Purpose**: Create target label, do some experiments

**Input**: data/processed/ochota_stops_with_amenities.csv

```python
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt
```

```python
PROJECT_DIR = Path.cwd().parent.parent
PROCESSED_CSV = PROJECT_DIR / "data" / "processed" / "ochota_stops_with_amenities.csv"
```

```python
df = pd.read_csv(PROCESSED_CSV)
df.head()
```

```python
from src.config import AMENITY_TAGS
for tag in AMENITY_TAGS:
    print(tag)
```

```python
df['target'] = 0

essential_cols = [f"{tag}_count" for tag in AMENITY_TAGS]
df["target"] = 1
for count_col in essential_cols:
    df.loc[df[count_col] < 5, "target"] = 0
```

```python
print(f"There are {len(df[df['target'] == 0])} labels '0' and {len(df[df['target'] == 1])} labels '1'.")
```
