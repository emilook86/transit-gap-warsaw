# OSM Data Collection Experiments

**Purpose**: OSMNX workflow investigation

**Input**: None (queries OpenStreetMap directly)

**Output**: Brief examples of Ochota buildings

```python
import osmnx as ox
import numpy as np
import time
from src.config import OCHOTA_BOUNDS, AMENITY_TAGS
```

```python
lat = np.mean([OCHOTA_BOUNDS["min_lat"], OCHOTA_BOUNDS["max_lat"]])
lon = np.mean([OCHOTA_BOUNDS["min_lon"], OCHOTA_BOUNDS["max_lon"]])
tags = AMENITY_TAGS["hospital_related"]
```

Now we query pharmacies and hospitals in the center of Ochota, as an example.

```python
start = time.time()

try:
    health_buildings = ox.features_from_point((lat, lon), tags=tags, dist=800)
    elapsed = time.time() - start
    print(
        f"Found {len(health_buildings)} pharmacies and hospitals in {elapsed:.3f} seconds"
    )
    display(health_buildings.head())
except Exception as e:
    print(f"Error: {e}")
```

Querying may fail if we have large distance circle to cover (or if we query many building types)
